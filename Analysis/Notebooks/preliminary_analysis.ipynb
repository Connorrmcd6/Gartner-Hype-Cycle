```python
from scipy.stats import norm
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.axisartist.axislines import SubplotZero
matplotlib.use("pgf")
matplotlib.rcParams.update({
"pgf.texsystem": "lualatex",
'font.family': 'serif',
'text.usetex': True,
'pgf.rcfonts': False,
})
matplotlib.rcParams["text.usetex"] = True
plt.rcParams["figure.figsize"] = [8, 4]
plt.rcParams.update({'font.size': 12})
plt.rcParams['figure.dpi'] = 300
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.axes.xaxis.set_ticklabels([])


mean = 5.3
std = 0.75

data = np.arange(0,10,0.01)
pdf = 100*norm.pdf(data , loc = mean , scale = std)
cdf = 100*norm.cdf(data, loc = mean)
 

plt.plot(data, pdf , color = 'black', label ='Adoption Rate')
plt.plot(data, cdf , color = '#E31B23', label ='Cumulative Adoption' )
plt.xlabel('Time')
plt.ylabel('Market Share (%)')
x_axis = np.arange(0, 10, 0.01)
plt.fill_between(x_axis, pdf, color= '#d6e6f4', where= (x_axis < 3.8), label = 'Innovators')
plt.fill_between(x_axis, pdf, color= '#abd0e6', where=(x_axis > 3.8) & (x_axis < 4.55), label = 'Early Adopters')
plt.fill_between(x_axis, pdf, color= '#6aaed6', where=(x_axis > 4.55) & (x_axis < 5.3), label = 'Early Majority')
plt.fill_between(x_axis, pdf, color= '#3787c0', where=(x_axis > 5.3) & (x_axis < 6.05), label = 'Late Majority')
plt.fill_between(x_axis, pdf, color= '#105ba4', where= (x_axis > 6.05), label = 'Laggards')
plt.legend()
plt.savefig(os.path.join(path, 'rogers_distribution.pdf'), format='pdf',bbox_inches='tight',pad_inches = 0)
plt.close()
```

```python
from datetime import datetime
#convert dates to correct format
def format1(d):
    return datetime.strptime(d, '%y-%b').strftime('%Y-%m')

def format2(d):
    return datetime.strptime(d, '%Y/%m/%d').strftime('%Y-%m-%d')

def format3(d):
    return datetime.strptime(d, '%b-%y').strftime('%Y-%m')
```

```python
ml = pd.read_csv('/Users/connormcdonald/Desktop/Masters/MIT807/Gartner Repository/Analysis/google_analytics_data/ML.csv')
blockchain = pd.read_csv('/Users/connormcdonald/Desktop/Masters/MIT807/Gartner Repository/Analysis/google_analytics_data/blockchain.csv')
covid = pd.read_csv('/Users/connormcdonald/Desktop/Masters/MIT807/Gartner Repository/Analysis/google_analytics_data/covid.csv')
dprint = pd.read_csv('/Users/connormcdonald/Desktop/Masters/MIT807/Gartner Repository/Analysis/google_analytics_data/3dprint.csv')
five = pd.read_csv('/Users/connormcdonald/Desktop/Masters/MIT807/Gartner Repository/Analysis/google_analytics_data/5g.csv')
ev = pd.read_csv('/Users/connormcdonald/Desktop/Masters/MIT807/Gartner Repository/Analysis/google_analytics_data/ev.csv')
iphone = pd.read_csv('/Users/connormcdonald/Desktop/Masters/MIT807/Gartner Repository/Analysis/google_analytics_data/iphone.csv')
home = pd.read_csv('/Users/connormcdonald/Desktop/Masters/MIT807/Gartner Repository/Analysis/google_analytics_data/smarthome.csv')
stream = pd.read_csv('/Users/connormcdonald/Desktop/Masters/MIT807/Gartner Repository/Analysis/google_analytics_data/stream.csv')



ml['date']  = ml['date'].apply(format1)
blockchain['date']  = blockchain['date'].apply(format1)
covid['date']  = covid['date'].apply(format2)
dprint['date']  = dprint['date'].apply(format3)
five['date']  = five['date'].apply(format1)
ev['date']  = ev['date'].apply(format3)
iphone['date']  = iphone['date'].apply(format3)
home['date']  = home['date'].apply(format3)
stream['date']  = stream['date'].apply(format3)


ml['date']  = pd.to_datetime(ml['date'])
blockchain['date']  = pd.to_datetime(blockchain['date'])
covid['date']  = pd.to_datetime(covid['date'])
dprint['date']  = pd.to_datetime(dprint['date'])
five['date']  = pd.to_datetime(five['date'])
ev['date']  = pd.to_datetime(ev['date'])
iphone['date']  = pd.to_datetime(iphone['date'])
home['date']  = pd.to_datetime(home['date'])
stream['date']  = pd.to_datetime(stream['date'])
```
